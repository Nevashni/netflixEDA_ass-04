```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Netflix Dataset.csv')
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
df["director"].fillna("Unknown", inplace=True)
```

```python
print(df['director'].isnull().sum())
```

```python
df["cast"].fillna("Unknown", inplace=True)
```

```python
print(df['cast'].isnull().sum())
```

```python
df["country"].fillna(df["country"].mode()[0], inplace=True)
```

```python
print(df['country'].isnull().sum())
```

```python
df["rating"].fillna(df["rating"].mode()[0], inplace=True)
```

```python
df["movie_duration"] = df["duration"].str.extract(r"(\d+) min")[0].astype(float)
```

```python
df["num_seasons"] = df["duration"].str.extract(r"(\d+) Season")[0].astype(float)
```

```python
df["movie_duration"].fillna(0, inplace=True)
df["num_seasons"].fillna(0, inplace=True)
```

```python
df.drop(columns=["duration"], inplace=True)
```

```python
df
```

```python
df_exploded = df.assign(listed_in=df["listed_in"].str.split(", ")).explode("listed_in")
```

```python
df_exploded
```

```python
category_counts = df_exploded["listed_in"].value_counts()

print(category_counts)
```

```python
df_exploded = df_exploded.drop_duplicates()
```

```python
df_exploded
```

```python
df_exploded["listed_in"] = df_exploded["listed_in"].str.strip().str.lower()
```

```python
category_mapping = {
    "international movies": "international",
    "international tv shows": "international",

    "tv dramas": "dramas",
    "romantic tv shows": "romance",
    "romantic movies": "romance",
    "dramas": "dramas",

    "tv comedies": "comedies",
    "comedies": "comedies",

    "horror movies": "horror",
    "tv horror": "horror",

    "thrillers": "thriller",
    "tv thrillers": "thriller",

    "crime tv shows": "crime",

    "documentaries": "documentary",
    "docuseries": "documentary",
    "science & nature tv": "documentary",

    "kids' tv": "kids & family",
    "children & family movies": "kids & family",

    "stand-up comedy & talk shows": "stand-up comedy",
    "stand-up comedy": "stand-up comedy",

    "action & adventure": "action",
    "tv action & adventure": "action",

    "sci-fi & fantasy": "sci-fi",
    "tv sci-fi & fantasy": "sci-fi",

    "reality tv": "reality",

    "british tv shows": "international",
    "spanish-language tv shows": "international",
    "korean tv shows": "international",

    "anime series": "anime",
    "anime features": "anime",

    "independent movies": "independent",

    "classic movies": "classic",
    "classic & cult tv": "classic",
    "cult movies": "classic",

    "faith & spirituality": "spirituality",

    "sports movies": "sports",

    "lgbtq movies": "lgbtq",

    "teen tv shows": "teen",

    "movies": "general movies",
    "tv shows": "general tv shows",
    "tv mysteries" : "mystery",

}
df_exploded["listed_in"] = df_exploded["listed_in"].replace(category_mapping)

df_exploded
```

```python
category_counts = df_exploded["listed_in"].value_counts()
print(category_counts)
```

```python
threshold = 30
rare_categories = category_counts[category_counts < threshold].index
df_exploded["listed_in"] = df_exploded["listed_in"].replace(rare_categories, "Other")
print(df_exploded["listed_in"].value_counts())
```

```python
df_exploded.drop_duplicates(subset=["show_id"], keep="first", inplace = True)
```

```python
df_exploded.rename(columns={"listed_in": "genre"}, inplace=True)
df_exploded
```

```python
df_exploded["rating"].replace({"74 min": "NR", "84 min": "NR","66 min": "NR"}, inplace=True)
```

```python
df_exploded['rating'].value_counts()
```

```python
df_exploded
```

```python
df_exploded[["actor 1", "actor 2", "actor 3"]] = df_exploded["cast"].str.split(", ", expand=True).iloc[:, :3]
```

```python
df_exploded
```

```python
df_exploded.drop(columns=["cast"], inplace=True)
```

```python
df_exploded
```

```python
df_exploded[["actor 1", "actor 2", "actor 3"]] = df_exploded[["actor 1", "actor 2", "actor 3"]].fillna("Unknown")
```

```python
df_exploded
```

```python
df_exploded[df_exploded["date_added"].isnull()]
```

```python
df_exploded["date_added"] = pd.to_datetime(df_exploded["date_added"], errors="coerce")
```

```python
df_exploded[df_exploded["date_added"].isnull()]
```

```python
df_exploded.info()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df_exploded, x='type', palette='pastel')
plt.title("Distribution of Content Type (Movies vs. TV Shows)")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_exploded['release_year'], bins=20, kde=True, color="blue")
plt.title("Trend of Movie Releases Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
top_countries=df_exploded['country'].value_counts().head(10)
plt.figure(figsize=(12, 5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xticks(rotation=45)
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df_exploded, x='rating', order=df_exploded['rating'].value_counts().index, palette='magma')
plt.xticks(rotation=45)
plt.title("Distribution of Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_genres=df_exploded['genre'].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_genres.values, labels=top_genres.index, autopct='%1.1f%%', colors=sns.color_palette('pastel', len(top_genres)))
plt.title("Top 10 Popular Genres Distribution")
plt.show()
```

```python
top_directors = df_exploded[df_exploded['director'] != 'Unknown']['director'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_directors.index, x=top_directors.values, palette='viridis')
plt.title("Top 10 Most Featured Directors")
plt.xlabel("Count")
plt.ylabel("Director")
plt.show()
```

```python
df_filtered = df_exploded[(df_exploded['actor 1'] != "Unknown") &
                          (df_exploded['actor 2'] != "Unknown") &
                          (df_exploded['actor 3'] != "Unknown")]
df_filtered['actor_combo'] = df_filtered[['actor 1', 'actor 2', 'actor 3']].agg(', '.join, axis=1)
top_actor_combos = df_filtered['actor_combo'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=top_actor_combos.index, x=top_actor_combos.values, palette='coolwarm')
plt.title("Top 10 Most Common Actor Combinations", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Actor Combination", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
```

```python
for col in ['type', 'director', 'country', 'rating', 'genre']:
    print(f"Unique values in {col}: {df_exploded[col].nunique()}")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=df_exploded['num_seasons'].dropna(), palette='coolwarm')
plt.title("Number of Seasons in TV Shows", fontsize=14)
plt.xlabel("Number of Seasons", fontsize=12)
plt.ylabel("Count of Shows", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_exploded['movie_duration'].dropna(), bins=30, kde=True, color='royalblue')
plt.title("Distribution of Movie Durations", fontsize=14)
plt.xlabel("Duration (Minutes)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
